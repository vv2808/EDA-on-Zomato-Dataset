```python
#4)Visualizing the locations as per the order.
plt.figure(figsize = [10, 6])

sns.countplot(x = df['listed_in(city)']) 
sns.countplot(x = df['listed_in(city)']).set_xticklabels(sns.countplot(x = df['listed_in(city)']).get_xticklabels(), rotation=90, ha= 'right')

plt.title('Location')
plt.ylabel('Number of Orders')
```

```python
#7)Visualize Online Order Facility in the Location Wise.
plt.figure(figsize = [20, 20])

sns.countplot(x = df['location']) 
sns.countplot(x = df['online_order']).set_xticklabels(sns.countplot(x = df['location']).get_xticklabels(), rotation=90, ha= 'right')

plt.title('Location wise online orders')
plt.xlabel('locations')
plt.ylabel('Number of online Orders')
```

```python
df = df.loc[df.rate !='NEW']
df = df.loc[df.rate != '-']
```

```python
df['rate'] = df['rate'].apply(lambda x: x.replace('/5',''))
df['rate'] = df['rate'].astype('float')
```

```python
#8)Visualize Restaurant VS Rate
bins = list(range(0, 6, 1))
np.random.seed(31415)

df_test = pd.DataFrame({'rate': df['rate'].astype(float), 
                        'rate_group': pd.cut(df['rate'].astype(float), bins=bins, right=False)})

ax = sns.countplot(data=df_test, x='rate_group')
ax.tick_params(axis='x', labelrotation=90)
plt.title('restaurant-rate')
plt.xlabel('rating range')
plt.ylabel('Restaurant count')
```

```python
#9) Compare the outliers(in this case i did for top 100)
plt.figure(figsize=(20,30))
famous_restaurant = df['name'].value_counts()[:100]
sns.barplot(x = famous_restaurant,y = famous_restaurant.index)
plt.title("Restaurants chains in Bangaluru")
plt.xlabel("Number of outliers")
plt.show()
```
